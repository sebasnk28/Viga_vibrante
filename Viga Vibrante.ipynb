```python
import numpy as np
from matplotlib import pyplot as plt
from numpy import pi

#condiciones iniciales y constantes
L=1
N=25

#generar la matriz de coeficientes
dp=(-6)*np.ones(N-1)
ds=(4)*np.ones(N-2)
dt1=(-1)*np.ones(N-3)
A = np.diag(dp,k=0)+ np.diag(ds,k=1)+ np.diag(ds,k=-1)+ np.diag(dt1,k=2)+ np.diag(dt1,k=-2)
A[0][0]=-5
A[-1][-1]=-5
```

Condiciones Iniciales

```python
def ci1(x):
    return np.sin(np.pi*x)
def ci2(x):
    return 0*x

dx=0.04
dt=0.000001
s=(dt/(dx**2))

p=0
m=p*1000000
x = np.arange(0.04,1,0.04)


u_0=[]
for i in x:
    u_0.append(ci1(i))  
u_1=[]
for r in x:
    u_1.append(ci1(r)+dt*ci2(r))


for i in range(1,m):
    u_2 = (s**2)*np.dot(A,u_0)- u_0 + np.dot(2,u_1)
    u_0=u_1
    u_1=u_2
    

#parte analítica
def analitica(x):
    t=3
    fx=lambda x: np.cos((pi**2)*t)*(np.sin(pi*x)) 
    sol=fx(x)
    return(sol)
xs=np.linspace(0,1,100)
an=analitica(xs)


xf=list(u_1)
xf1=[0]
for k in xf:
    xf1.append(k)

xf1.append(0)   
xs1 = np.arange(0,1+dx,0.04) 

plt.plot(xs1,xf1,label='Numérica')
#plt.plot(xs,an,label='Analítica')       
plt.title('Solución Ecuación , t={0}'.format(p))
plt.grid()
plt.legend()
plt.show()
```
